```python
df[df['Predicted_Leave_Status']==0]['Contract'].value_counts()
df[df['Predicted_Leave_Status']==0]['Contract'].plot
```

```python
# Filter the dataframe where Predicted_Leave_Status is 0
filtered_df = df[df['Predicted_Leave_Status'] == 0]

# Get the counts of unique 'Contract' values
contract_counts = filtered_df['Contract'].value_counts()

# Plot the pie chart
contract_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))

# Display the plot
plt.ylabel('')  # This removes the y-label that is automatically added
plt.title('Distribution of Contract for Predicted_Leave_Status = 0')
plt.show()
```
