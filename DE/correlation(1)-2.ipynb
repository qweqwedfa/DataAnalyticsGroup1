```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    'tenure': list(range(1, 73)),  # Example tenure from 1 to 72 months
    'Churn': [0.2] * 36 + [0.3] * 12 + [0.1] * 24  # Dummy churn data
})


churn_by_tenure = df.groupby('tenure')['Churn'].mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(churn_by_tenure.index, churn_by_tenure.values, marker='o')
plt.xlabel('Tenure (Months)')
plt.ylabel('Churn Rate')
plt.title('Churn Rate Over Tenure')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd


df = pd.read_csv('/correlation/GROUP1_TELCO_CHURN_for_Correlation(in).csv')
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['1-12', '13-24', '25-36', '37-48', '49-60', '61-72']


df['tenure_range'] = pd.cut(df['tenure'], bins=bins, labels=labels)


churn_by_tenure = df.groupby('tenure_range')['Churn'].mean()


import matplotlib.pyplot as plt
churn_by_tenure.plot(kind='bar', title='Churn Rate by Tenure Range')
plt.ylabel('Churn Rate')
plt.xlabel('Tenure Range (Months)')
plt.show()
```
